```python
# 1. Create two numpy array and plot using matplotlib.
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
plt.plot(x,y,label="Line",marker="*",markersize=15,markerfacecolor="red",
         markeredgecolor="black",linestyle="-.",linewidth=2,alpha=0.4)
plt.legend(loc="lower right",shadow=True,bbox_to_anchor=(0.65,0.98),
           labelcolor="magenta",edgecolor="green",facecolor="lightblue")
for i , (xi,yi) in enumerate(zip(x,y)):
    plt.annotate(f"({xi},{yi})",(xi,yi),ha="center",va="bottom")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.yticks(fontsize=13)
plt.title("Lineplot",rotation=230)
plt.grid(True,axis="y",linewidth=2,color="green",linestyle=":")
plt.show()
```

```python
# 2. What is bubble chart? Show it’s plotting.
""" Bubble Chart is used to look at relationships between three numeric variables."""
x=[1,2,3,4,5]
y=[5,10,23,17,7]
sizes=[30,45,55,65,75]
plt.scatter(x,y,s=sizes,color=["blue","red","green","yellow","magenta"]
            ,alpha=0.7,marker="D",edgecolor="black",linewidths=3)
plt.show()
```

```python
# What is a bubble chart, and how is it different from a scatter plot? Also show plotting for the same.
"""Representation of Data:
Scatter Plot: In a scatter plot, each data point is represented by a single marker (usually a dot) all having same size.
Bubble Chart: In a bubble chart, each data point is represented by a bubble (circle) whose size indicates the value of a third variable associated with that data point.

Visualization of Three Variables:
Scatter Plot: A scatter plot visualizes the relationship between two variables (typically plotted on the x and y axes) but does not directly represent a third variable.
Bubble Chart: A bubble chart visualizes the relationship between two variables on the x and y axes, similar to a scatter plot, but also incorporates a third variable through the size of the bubbles."""
import numpy as np
x=np.random.randn(100)
y=np.random.randn(100)
color=np.random.randn(100)
plt.scatter(x,y,c=color,cmap="terrain",s=50)
plt.colorbar(label="Values")
plt.show()
```

```python
# Write a code for visualizing non-uniform 2D data using scatter plotting. Also add colormap, title, x-label and y-label.
import numpy as np
x=np.random.randn(100)
y=np.random.randn(100)
color=np.random.randn(100)
plt.scatter(x,y,c=color,cmap="viridis",s=50)
plt.colorbar(label="Values")
plt.xlabel("X-AXis")
plt.ylabel("Y-AXis")
plt.title("Non- uniform 2D graph")
plt.show()
```

```python
# Show the plotting command line with minimum three attributes for given 5 plots:

# 1. Waffle Chart
import matplotlib.pyplot as plt
from pywaffle import Waffle
import pandas as pd
df=pd.DataFrame({"phones":["samsung","xiaomi","apple","lenovo"],"usage":[23,15,45,24]})
plt.figure(FigureClass=Waffle,values=df.usage,icons="phone",rows=5,icon_size=25,
           labels=list(df["phones"]),facecolor="lightyellow",colors=["Red","cyan","purple","maroon"])
plt.show()
```

```python
# 2. Histogram
import numpy as np
import matplotlib.pyplot as plt
x=np.random.rand(50)
plt.hist(x,bins=7,color="skyblue",alpha=0.222,orientation="horizontal",histtype="stepfilled",edgecolor="black")
plt.show()
```

```python
# 3. Area Plot
x=[1,2,3,4,5]
y1=[3,6,9,14,12]
y2=[12,14,8,9,3]
plt.fill_between(x,y1,y2,interpolate=False,step="mid",color="yellow",edgecolor
                 ="black",alpha=0.5,linestyle=":",linewidth=4,label="area")
plt.legend()
plt.show()
```

```python
# 4.Polar plot
theta=np.linspace(0,2*np.pi,7)
radius=[3,6,2,5,1,4,7]
plt.polar(theta,radius,marker="*",color="magenta",markersize=4,linestyle="-.",
          markerfacecolor="red",markeredgecolor="black",label="Magnitude")
plt.legend()
plt.show()
```

```python
# 5. Box Plot
x=[[10,15,20,25,30],
[25,30,35,40,45],
[5,10,15,20,25]]
plt.boxplot(x,patch_artist=True,notch=True,vert=True,labels=["i1","i2","i3"]
            ,boxprops={"color":"pink"},widths=0.5)
plt.show()
```

```python
# Write a code and plot bar plots and pie charts using the following categories and values: 
# Categories = ['Category A', 'Category B', 'Category C', 'Category D']
# values = [15, 30, 35, 20].
Categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 30, 35, 20]
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.bar(Categories,values,color=["lightgreen","cyan","yellow","orange"],edgecolor="black")

axes2=fig.add_axes([1,0,1,1])
axes2.pie(values,labels=Categories,autopct="%1.2f%%",colors=["lightgreen","cyan","yellow","orange"]
          ,startangle=120,textprops=dict(color="magenta"),wedgeprops={"edgecolor":"black","linewidth":0.5},shadow=True)
plt.show()
```

```python
# Show the contour plot for function: Z = np.sin(X) + np.cos(Y).
x=np.linspace(-6,6,20)
y=np.linspace(-6,6,20)
X,Y=np.meshgrid(x,y)
Z= np.sin(X) + np.cos(Y)
plt.contour(X,Y,Z)
plt.show()
```

```python
# Write a code to generate 3D line plot as well as scatter plot and show visualization output. 
# 3d line plot
z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)
ax=plt.axes(projection="3d")
ax.plot3D(x,y,z,color="green")
plt.show()

# Scatter 3D plot
z=np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)
color=np.random.randn(100)
ax=plt.axes(projection="3d")
ax.scatter(x,y,z,c=color)
plt.show()
```
